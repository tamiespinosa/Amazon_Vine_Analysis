```python
# Dependencies and Setup
import pandas as pd
import os
```

```python
# Read the Vine Table Data and Student Data and store into a Pandas DataFrame
vine_df = pd.read_csv("Resources/vine_table.csv")
vine_df
```

```python
vine_20df = vine_df[vine_df["total_votes"]>20]
vine_20df
```

```python
vine_20df["percent_helpful_votes"] = (vine_20df["helpful_votes"]/vine_20df["total_votes"])*100
vine_20df
```

```python
vine_50df = vine_20df[vine_20df["percent_helpful_votes"]>=50]
vine_50df 
```

```python
vine_paid = vine_50df[vine_50df["vine"]=="Y"]
vine_paid
```

```python
vine_unpaid = vine_50df[vine_50df["vine"]=="N"]
vine_unpaid
```

```python
paid_count = vine_paid["review_id"].count()
paid_count_5 = vine_paid[vine_paid["star_rating"]==5.0]["review_id"].count()
percent_paid_count_5 = (paid_count_5/paid_count)*100
print(paid_count, paid_count_5,percent_paid_count_5)
```

```python
unpaid_count = vine_unpaid["review_id"].count()
unpaid_count_5 = vine_unpaid[vine_unpaid["star_rating"]==5.0]["review_id"].count()
percent_unpaid_count_5 = (unpaid_count_5/unpaid_count)*100
print(unpaid_count, unpaid_count_5,percent_unpaid_count_5)
```
